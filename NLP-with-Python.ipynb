```python
def get_ratio(row):
    name = row['Expedia']
    name1 = row['Booking.com']
    return fuzz.token_set_ratio(name, name1)
df[df.apply(get_ratio, axis=1) > 70].head(10)
```

```python
len(df[df.apply(get_ratio, axis=1) > 70]) / len(df)
```
